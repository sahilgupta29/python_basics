Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

```python
#Answer
""" Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in a wide range 
of formats. It provides a flexible and user-friendly interface for generating various types of plots and charts. 

Matplotlib is widely used in fields such as data analysis, scientific research, engineering, and more, to effectively communicate data and insights through visual representations.

The `pyplot` module of Matplotlib is a collection of functions that provides a MATLAB-like interface for creating plots. It simplifies the process of creating various types of plots. Here are five types of plots that can be created using the `pyplot` module of Matplotlib:

1. Line Plot: A basic plot where data points are connected by straight lines, commonly used to visualize trends over time or across categories.

2. Scatter Plot: Used to display the relationship between two variables by plotting individual data points as dots on a graph.

3. Bar Plot: Displays data using rectangular bars, with one axis representing categories and the other axis representing values.

4. Histogram: Represents the distribution of a dataset by dividing it into bins and displaying the frequency or proportion of data points in each bin.

5. Pie Chart: Represents data as sectors of a circular pie, with each sector's angle proportional to the value it represents. Often used to display parts of a whole.

These are just a few examples, and Matplotlib offers a wide range of plot types and customization options to suit various visualization needs."""
```

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
"""A scatter plot is a type of data visualization that uses dots to represent the values obtained for two different variables in
a dataset. Each dot in the plot represents a single data point, and its position on the graph corresponds to the values of the
two variables for that data point. Scatter plots are useful for identifying patterns, relationships, and trends between two 
continuous variables."""
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph")
plt.scatter(x,y,c="b")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
"""The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It is especially useful 
when you want to display multiple plots side by side or in a grid pattern. This can help you compare different datasets, 
visualize various aspects of your data, or present related information together."""
import numpy as np
import matplotlib.pyplot as plt

# Data for the plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1, marker='o')
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, marker='o')
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, marker='o')
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, marker='o')
plt.title('Line 4')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
#Answer
"""A bar plot is a graphical representation of data using rectangular bars, where the lengths of the bars are proportional to
the values they represent. Bar plots are used to compare and display categorical data or to show the relationship between 
different groups. They are particularly useful for visualizing discrete data and making comparisons between different 
categories."""
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profit")



```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
#Answer
"""A box plot, also known as a box-and-whisker plot, is a statistical visualization tool that provides a summary of the
distribution of a dataset. It displays the five-number summary of a dataset: the minimum, first quartile (25th percentile),
median (50th percentile), third quartile (75th percentile), and maximum. The box plot consists of a rectangular "box" that 
spans from the first quartile to the third quartile, with a vertical line inside representing the median. "Whiskers" extend
from the box to the minimum and maximum values that are within a certain range (often defined by 1.5 times the interquartile 
range), and any data points outside this range are considered outliers and marked individually."""
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
plt.boxplot(data, labels=labels)
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot Example')
plt.tight_layout()
plt.show()
```
